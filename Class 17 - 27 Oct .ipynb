```python
# Question
# Take a string as input.
# Convert the string to lowercase without using any inbuilt function.

# "This is A stRIng" > 'this is a string'
```

```python
for i in range(ord('a'), ord('z')+1):
    print(i, chr(i))
for i in range(ord('A'), ord('Z')+1):
    print(i, chr(i))
```

```python
# 1 loop over all the chars
# 2 find if the char is capital/upper OR not
# 3 "a" >> "A": ascii+ 32>> lowercase
# 4 create new str with the modified chars
```

```python
def convert2lowercase(my_str):
    result = ""
    for char in my_str:# 1 loop over all the chars
        if ord(char)>=ord('A') and ord(char)<=ord('Z'):# 2 find if the char is capital/upper OR not
            lowerChar = chr(ord(char)+32) # 3 "a" >> "A": ascii+ 32>> lowercase
        else:
            lowerChar= char
        result+=lowerChar
    return result
convert2lowercase(my_str="This is A stRIng")
```

```python
# ABC---XYZ
# abc---xyc
def convert2lowercase(my_str):
    for i in range(ord('A'), ord('Z')+1):
        my_str=my_str.replace(chr(i), chr(i+32))
    return my_str
convert2lowercase(my_str="This is A stRIng")
```

```python
# question
# Take a string as input.
# Replace all the space with underscore.

```

```python
def change2underscore(my_str):
    result = ""
    for i in my_str:
        if i.isspace():
            result+='_'
        else:
            result+=i
    return result

change2underscore(my_str="This is A stRIng")
```
